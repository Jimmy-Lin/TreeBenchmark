# Scalability
Measurements of how performance varies across different worker pool sizes

## Generate Dataset
Use `python/scalability.py` to run the data collection

## Load Dataset

```python
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import pandas as pd
import numpy as np

dataset = pd.DataFrame(pd.read_csv('experiments/analysis/scalability/compas.csv', delimiter=","))
```

## Multi-line Plot

```python
max_dimensions = max(dataset['dimensions'])
max_workers = max(dataset['workers'])

fig = plt.figure(figsize=(10, 4), dpi=100)
for j in range(1, max_dimensions + 1):
    subset = dataset[dataset['dimensions'] == j]
    time = subset['time']
    workers = subset['workers']
    plt.plot(workers, time, label="{} dimensions".format(j), markersize=2, marker='o', linewidth=1)
plt.xlabel('Number of Workers')
plt.ylabel('Runtime (s)')
plt.grid()
plt.legend()
plt.title('Runtime vs Number of Workers')

fig = plt.figure(figsize=(10, 4), dpi=100)
for j in range(1, max_dimensions + 1):
    subset = dataset[dataset['dimensions'] == j]
    reference_speed = dataset[dataset['dimensions'] == j][dataset['workers'] == 1]['time'].iloc[0]
    speedup = reference_speed / subset['time']
    workers = subset['workers']
    plt.plot(workers, speedup, label="{} dimensions".format(j), markersize=2, marker='o', linewidth=1)
plt.xlabel('Number of Workers')
plt.ylabel('Speed-Up')
plt.grid()
plt.legend()
plt.title('Speed-Up vs Number of Workers')
```
