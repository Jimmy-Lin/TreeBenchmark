## Regularization Trade-Off
Measurements of how training and test accuracy varies at different levels of regularization and resulting complexity

## Generate Dataset
Use to `python/regularization.py` to run the data collection

## Load Dataset

```python
import matplotlib.pyplot as plt
import pandas as pd

dataset = pd.DataFrame(pd.read_csv('experiments/analysis/regularization/compas.csv', delimiter=","))
```

## Accuracy Trade-Off

```python
from math import log, exp

regularization = dataset['regularization']
ln_regularization = dataset['regularization'].apply(lambda x : -log(x))
leaves = dataset['leaves']
ln_leaves = dataset['leaves'].apply(lambda x : log(x))
training_accuracy = dataset['training_accuracy']
test_accuracy = dataset['test_accuracy']

fig = plt.figure(figsize=(10, 8), dpi=100)
plt.plot(leaves, training_accuracy, label='training_accuracy', markersize=8, marker='o', linewidth=0)
plt.plot(leaves, test_accuracy, label='test_accuracy', markersize=8, marker='o', linewidth=0)
plt.xlabel('Number of Leaves')
plt.ylabel('Training and Test Accuracy')
plt.grid()
plt.legend()
plt.title('Training and Test Accuracy vs Number of Leaves')

fig = plt.figure(figsize=(10, 8), dpi=100)
plt.plot(ln_leaves, training_accuracy, label='training_accuracy', markersize=8, marker='o', linewidth=0)
plt.plot(ln_leaves, test_accuracy, label='test_accuracy', markersize=8, marker='o', linewidth=0)
plt.xlabel('ln(Number of Leaves)')
plt.ylabel('Training and Test Accuracy')
plt.grid()
plt.legend()
plt.title('Training and Test Accuracy vs ln(Number of Leaves)')

fig = plt.figure(figsize=(10, 8), dpi=100)
plt.plot(ln_regularization, training_accuracy, label='training_accuracy', markersize=8, marker='o', linewidth=0.5)
plt.plot(ln_regularization, test_accuracy, label='test_accuracy', markersize=8, marker='o', linewidth=0.5)
plt.xlabel('-ln(Regularization)')
plt.ylabel('Training and Test Accuracy')
plt.grid()
plt.legend()
plt.title('Training and Test Accuracy vs -ln(Regularization)')
```
